```python
# 线条和颜色
plt.plot([1,2,3,4,5],[1,4,9,16,25],':',color='r')
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
```

```python
# 线条宽度
x = np.linspace(-10, 10)
y = np.sin(x)
plt.plot(x, y, linewidth=2.0)
```

```python
# 添加注释
plt.plot(x,y,color='b',linestyle=':',marker='o',
         markerfacecolor='r',markersize=5
         )
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.title('simple')
plt.text(0,0,'a1')
plt.grid(True)
plt.annotate('x2',xy=(-5,0),xytext=(-2,0.3),arrowprops=dict(facecolor='black',shrink=0.5))
```

```python
# 多个图
# 211 表示两行一列 第一个图
plt.subplot(211)
plt.plot(x, y,color='r')

plt.subplot(212)
plt.plot(x, y, color='b')

```

```python
x = range(10)
y = range(10)

fig = plt.gca()
plt.plot(x,y)

fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

```python
x = np.random.normal(loc=0.0,scale=1.0,size=300)
width=0.5
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)
ax = plt.subplot(111)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(bottom='off',top='off',left='off',right='off')
plt.grid()
plt.hist(x,alpha=0.5,bins=bins)
```
